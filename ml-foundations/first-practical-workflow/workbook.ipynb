## Step 02 - Create a tiny dataset

```python
import pandas as pd

columns = [
    "bedrooms",
    "floor_area_m2",
    "distance_station_km",
    "age_years",
    "price_k",
]
house_rows = [
    [1, 45, 0.5, 8, 210],
    [2, 60, 1.2, 30, 245],
    [2, 72, 0.8, 12, 280],
    [3, 80, 2.5, 45, 300],
    [3, 92, 1.0, 20, 335],
    [3, 105, 3.1, 5, 390],
    [4, 110, 1.8, 35, 410],
    [4, 130, 2.2, 15, 455],
    [4, 145, 4.5, 60, 430],
    [5, 155, 2.8, 10, 520],
    [5, 175, 5.0, 75, 500],
    [6, 205, 3.7, 25, 610],
]
houses = pd.DataFrame(house_rows, columns=columns)

houses
```

```python
print("houses shape:", houses.shape)
print("columns:", list(houses.columns))
```

```python
feature_columns = [
    "bedrooms",
    "floor_area_m2",
    "distance_station_km",
    "age_years",
]
X = houses[feature_columns]
y = houses["price_k"]
```

```python
print("First feature rows:")
print(X.head())
print()
print("First target values:")
print(y.head())
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
print("Same number of rows:", len(X) == len(y))
print()
print("First feature row:")
print(X.iloc[0])
print()
print("Matching target value:")
print(y.iloc[0])
```

## Step 03 - Before Training

```python
from sklearn.model_selection import train_test_split
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=42,
)
```

```python
print("X_train:", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train:", y_train.shape)
print("y_test: ", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
house_model = LinearRegression()
```

```python
house_model.fit(X_train, y_train)
```

```python
coefficients = pd.Series(
    data=house_model.coef_,
    index=feature_columns,
    name="learned_weight",
)
print("Intercept:", round(house_model.intercept_, 2))
coefficients.round(2)
```
