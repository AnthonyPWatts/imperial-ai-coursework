## 04a — Matrix-Vector Multiplication

[Open notebook 04a](04a-matrix-vector-multiplication.ipynb)

This notebook introduces the core idea:

> a matrix can transform a vector into another vector.

The main calculation is:

$$
A v
$$

or, in NumPy:

```python
A @ v
```

This should establish the basic mental model:

> matrix = transformation rule

## 04b — How 2×2 Matrix Elements Work

[Open notebook 04b](04b-how-2x2-matrix-elements-work.ipynb)

This notebook looks inside a 2×2 matrix:

$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

and explains what each element contributes.

For:

$$
v = [x, y]
$$

the output is:

$$
A v = [ax + by, cx + dy]
$$

So:

- \(a\): input x contributes to output x
- \(b\): input y contributes to output x
- \(c\): input x contributes to output y
- \(d\): input y contributes to output y

This notebook is the bridge between “I can multiply a matrix by a vector” and “I can understand the shape of a transformation”.

## 04d — Rotation Transformations

[Open notebook 04d](04d-rotation-transformations.ipynb)

This notebook introduces rotation matrices.

The main 2D rotation matrix is:

$$
R_\theta =
\begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta)
\end{bmatrix}
$$

The key ideas are:

- rotation changes direction
- pure rotation preserves length
- positive angles rotate counter-clockwise
- sine and cosine describe the rotated coordinates

## 04e — Basis Vectors and Matrix Columns

[Open notebook 04e](04e-basis-vectors-and-matrix-columns.ipynb)

This is the conceptual wrap-up for the 04 section.

The most important idea is:

> the columns of a matrix show where the basis vectors go

For:

$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

the first column tells us where:

$$
e_1 = [1, 0]
$$

goes.

The second column tells us where:

$$
e_2 = [0, 1]
$$

goes.

Once you know where the basis vectors go, you know what the matrix does to the whole plane.

## Section summary

The main idea of this section is:

> matrices transform vectors

For 2D vectors, a 2×2 matrix can stretch, flip, shear or rotate the plane.

The most important long-term mental model is:

> a matrix is described by where it sends the basis vectors

That idea will matter later for:

- linear systems
- least squares
- regression
- dimensionality reduction
- neural network layers
- embeddings and feature transformations
