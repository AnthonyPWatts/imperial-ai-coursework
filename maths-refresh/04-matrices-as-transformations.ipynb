# 04 — Matrices as Transformations

## Purpose

This section introduces one of the most useful ways to understand matrices:

> a matrix can be read as a transformation of space.

For this section, we focus on 2D vectors and 2×2 matrices.

The aim is not to memorise lots of matrix formulas. The aim is to build a visual and geometric intuition for what matrices do.

## Prerequisites

This section assumes you are comfortable with the earlier vector ideas:

- a vector has components such as $[2, 1]$
- vectors can be added
- vectors can be scaled by numbers
- the dot product is a separate operation from matrix-vector multiplication

The most important idea to bring forward is that a vector can be read as a combination of basis directions.

## The central idea

The standard basis vectors in 2D are:

$$
\mathbf{e}_1 =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
\qquad
\mathbf{e}_2 =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

A vector with coordinates $x$ and $y$ can be read as:

$$
\mathbf{v} =
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
x\mathbf{e}_1 + y\mathbf{e}_2
$$

So a vector is not just an arrow.

It is also a set of instructions:

> take $x$ lots of $\mathbf{e}_1$, and $y$ lots of $\mathbf{e}_2$.

## What a matrix does

A 2×2 matrix moves the basis vectors.

For:

$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

the first column tells us where $\mathbf{e}_1$ lands:

$$
A\mathbf{e}_1 =
\begin{bmatrix}
a \\
c
\end{bmatrix}
$$

and the second column tells us where $\mathbf{e}_2$ lands:

$$
A\mathbf{e}_2 =
\begin{bmatrix}
b \\
d
\end{bmatrix}
$$

Once the basis vectors have moved, every other vector is rebuilt from those moved basis vectors:

$$
A\mathbf{v}
=
x(A\mathbf{e}_1) + y(A\mathbf{e}_2)
$$

That is the central mental model for the whole section:

> the coefficients stay the same; the basis vectors change.

## 04a — Matrix-vector multiplication

The first notebook introduces matrix-vector multiplication as more than a calculation.

For a vector:

$$
\mathbf{v} = 2\mathbf{e}_1 + 1\mathbf{e}_2
$$

matrix-vector multiplication gives:

$$
A\mathbf{v} = 2(A\mathbf{e}_1) + 1(A\mathbf{e}_2)
$$

The notebook shows that the matrix changes the basis vectors, and the vector follows because it is built from those basis vectors.

## 04b — How 2×2 matrix elements work

The second notebook opens up a general 2×2 matrix:

$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

It looks at the row view:

$$
A
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
ax + by \\
cx + dy
\end{bmatrix}
$$

and connects that to the column view:

$$
A\mathbf{e}_1 =
\begin{bmatrix}
a \\
c
\end{bmatrix}
\qquad
A\mathbf{e}_2 =
\begin{bmatrix}
b \\
d
\end{bmatrix}
$$

The rows explain the arithmetic.

The columns explain the geometry.

## 04c — Scaling, reflection and shear

The third notebook applies the column view to common named transformations.

Scaling changes basis-vector lengths.

Reflection flips one basis direction.

Shear slants one basis direction into the other.

The important point is that these are not separate magic tricks. They are all examples of the same matrix idea:

> move the basis vectors, then rebuild every other vector from them.

## 04d — Rotation transformations

The fourth notebook introduces rotation matrices.

A 2D rotation matrix has the form:

$$
R_\theta =
\begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta)
\end{bmatrix}
$$

The columns are the rotated basis vectors:

$$
R_\theta\mathbf{e}_1 =
\begin{bmatrix}
\cos(\theta) \\
\sin(\theta)
\end{bmatrix}
\qquad
R_\theta\mathbf{e}_2 =
\begin{bmatrix}
-\sin(\theta) \\
\cos(\theta)
\end{bmatrix}
$$

Rotation changes direction, but preserves length.

The transformed basis vectors stay length $1$ and remain at right angles.

## Section summary

By the end of this section, you should be able to read a 2×2 matrix in two complementary ways.

The **row view** tells you how the output coordinates are calculated:

$$
A
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
ax + by \\
cx + dy
\end{bmatrix}
$$

The **column view** tells you where the basis vectors land:

$$
A =
\begin{bmatrix}
| & | \\
A\mathbf{e}_1 & A\mathbf{e}_2 \\
| & |
\end{bmatrix}
$$

The column view is the geometric one.

It lets you see a matrix as a transformation of space.

## The core intuition to remember

A vector can be built from basis vectors:

$$
\mathbf{v}
=
x\mathbf{e}_1 + y\mathbf{e}_2
$$

A matrix moves the basis vectors:

$$
\mathbf{e}_1 \mapsto A\mathbf{e}_1
\qquad
\mathbf{e}_2 \mapsto A\mathbf{e}_2
$$

Then the transformed vector is rebuilt using the same coefficients:

$$
A\mathbf{v}
=
x(A\mathbf{e}_1) + y(A\mathbf{e}_2)
$$

That one idea explains the whole section.

## What comes next

The next natural step is to transform more than one vector at a time.

Instead of asking:

> where does this one vector go?

we can ask:

> what happens to a whole shape when every point is transformed?

That leads into transforming grids, squares, triangles, and other simple shapes with matrices.

It also prepares the ground for combining transformations, where order starts to matter.
